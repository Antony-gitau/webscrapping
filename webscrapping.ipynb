```python
#using pandas to scrapt a webpage

import pandas as pd
```

```python
data = pd.read_html('https://en.wikipedia.org/wiki/Programming_languages_used_in_most_popular_websites')
```

```python
#printing the data
data
```

```python
#create a dataframe using the first line of the data

df = data[0]
```

```python
df.head()
```

```python
#extracting popularity column

df['Popularity(unique visitors per month)[1]']
```

```python
#extracting websites

df['Websites']
```

```python
#we can also use dot column name for websites

df.Websites
```

```python
#EXTRACT backend server side

df['Back-end(Server-side)']

```

```python
#statistics 

df['Popularity(unique visitors per month)[1]'].describe()
```

```python
#visualize

import matplotlib.pyplot as plt
```

```python
websites = df.Websites
unique_visitors = df['Popularity(unique visitors per month)[1]']
```

```python
exp = [0.25,0.25,0.25,0.1,0.1,0.1,0.15,0.1,0.15,0.2,0.25,0.1,0.3]
plt.pie(unique_visitors ,labels=websites, shadow=True, explode=exp, radius=3)
plt.show()
```
